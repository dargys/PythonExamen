## 1. The Requirement

*The requirements has two dimensions:*
- *Fields: Required fields to be present and populated.*
- *The 10-day rule: the invoice must be issued within 10 calendar days of the confirmed delivery date.*
  
*Key fields include seller and buyer identification (company name, country, VAT registration number), financial data (net amount, VAT rate, VAT amount, total, currency), and administrative fields (invoice type, status, and for credit notes, a reference to the original invoice).*



## 2. Our Data

TechDist GmbH has provided 5 datasets from their ERP system representing the full transaction flow from customer registration through to invoicing. The tables connect as follows: customer master and delivery addresses define who we sell to and where. Orders reference both. Deliveries confirm each order. Invoices are issued against confirmed deliveries.



## 3. The Gap

**Missing Fields:**
>*<span style="color:red;">seller_vat_reg_no</span> : it is a constant, should be stored in the ERP system*<br>
>*<span style="color:red;">seller_iban</span> : it is a constant, should be stored in the ERP system*<br>
>*<span style="color:red;">buyer_vat_reg_no</span> : It partially exists; two are completely missing, and the rest need to be corroborated for accuracy*<br>
>*<span style="color:red;">reverse_charge</span> : The field is missing, it can gets from the vat_category_code=AE. AE is the code that means that the the buyer is going to be charge(cross-border)*<br>

**Fields that exist but have quality issues:**
>*<span style="color:red;">invoice_line_no</span> : it's not unique, it can be a composite key together with other field like invoice_id*<br>
>*<span style="color:red;">buyer_country</span> : inconsistent formats(DE / DEU / Deutschland all used for Germany)*<br>

**The 10 days rule:**
>In order to determine whether this rule is fulfilled, I need the confirmed delivery date from the delivery table. The values there are inconsistent, with different formats, although they all have in common that the day appears first. I have standardized the dates and performed the calculation; approximately one-third of the orders do not comply with the requirement.*<br>


## 4. The Impact

> *How widespread do you expect the problem to be?* <br> 
 
*The company is not prepared for ViDA cross border B2B invoicing.*

> *Which customers or transactions might be most at risk?* <br>

*The customers that doesn't have VAT number. All cross-border invoices for these customers would be non-compliant.* <br>

> *What would happen if TechDist GmbH had to send e-invoices today?*<br>

*Immediate rejection risk: if invoices were submitted today, they would be rejected due the missing mandatory fields.*<br>


## 5. Next Steps

> *What would you recommend fixing first — and why?*<br>

*I would start by completing the missing information — this is data the company should already have, but it is clearly not stored in the right place. Next, derive the reverse_charge field from the existing vat_category_code, since the logic is already there. After that, clean and normalize the format of each field to ensure consistency. Finally, set up a trigger after delivery confirmation so that invoices are always generated within the 10-day window.*<br>

> *Who in the business needs to be involved to fix each issue?*<br>  

**IT / ERP team:**<br>
*- Add seller constant fields (name, country, VAT, IBAN) to invoice output as static configuration*<br>
*- Standardize all date fields*<br>
*- Standardize country codes*<br>
*- Add reverse_charge as an explicit boolean field derived from vat_category_code*<br>
- Rename fields to match ViDA naming (line_amount → invoiced_amount_net, etc.)?<br>
- Fix invoice_line_no to be unique per invoice?.<br>

**Accounting:**<br>
*- Collect missing customer's VAT numbers.*<br>
*- Validate all existing VAT numbers.*<br>

**Operations:**<br>
*- Implement the 10-day trigger: invoice generation should be automatically initiated when a delivery confirmation is received in the ERP*<br>
  
> *What would "good" look like after the fixes are applied?*<br>

*After the fixes are applied, the system will be able to assemble and submit a complete, ViDA-compliant e-invoice — with all required fields populated, consistent formats, and invoices issued within the 10-day window.*
