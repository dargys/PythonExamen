### 1. Language Characteristics

Python is an **interpreted, high-level, general-purpose** programming language. Its core features include:

- **Dynamic Typing**: Variable types are determined at runtime based on the assigned value — no need to declare a type in advance. This improves development flexibility.
- **Interpreted Language**: Code is interpreted and executed line-by-line at runtime, skipping a compilation step and speeding up the development-test cycle.

### 3. Variables

Variables are created via assignment. There is no need to declare a type first.

Variables are assigned via `=`. The Python equal sign is `==`. See more about operators in section 9 below. 

Variables must be defined (assigned a value) before use, otherwise a `NameError` will occur.

````python
var1 = 1
var2 = 10

# Assign the same value to multiple variables
a = b = c = 1

# Assign multiple values at once (unpacking)
a, b, c = 1, 2, "Python"
print(a, b, c)
````

### 4. Strings & f-strings

Strings can use single or double quotes. 

For formatting values into strings, use **f-strings** (Python 3.6+): prefix the string with `f` and place any expression inside `{}`. It is an easy way to generate strings that contain interpolated expressions. Any code between curly braces {} will be evaluated and then the result will be turned into a string and intserted into the overall string. 

F-strings are really useful. Try with below examples. Read more at [W3schools.com](https://www.w3schools.com/python/python_string_formatting.asp) 

````python
product = "Headset"
revenue = 7576.725
growth = 0.0523

# Basic embedding
print(f"Product: {product}")            # Product: Headset

# Number formatting
print(f"Revenue: {revenue:,.2f}")       # Revenue: 7,576.73
print(f"Growth:  {growth:.1%}")         # Growth:  5.2%

# Any expression works inside {}
print(f"Double revenue: {revenue * 2:,.0f}")    # Double revenue: 15,153
````

### 5. Operators

Python supports the following categories of operators:

| Category | Examples |
|---|---|
| Arithmetic | `+`, `-`, `*`, `/`, `//`, `%`, `**` |
| Comparison | `==`, `!=`, `>`, `<`, `>=`, `<=` |
| Logical | `and`, `or`, `not` |

Note: `/` always returns a float; `//` is floor (integer) division.

### 6. Conditional Statements & Control Flow

Python by default evaluates every line of code from top to bottom, sequentially. The order in which Python runs our code is called **control flow**, and Python provides functionality we can use to change the control flow. 

The first tool Python provides for us is the **if-statement**. 

We write an [**if-statement**](https://www.w3schools.com/python/gloss_python_if_statement.asp) like this:

```python
    if bool:
        code_1
    else:
        code_2
```

`bool` here is a boolean expression, evaluating to either **`True`** or **`False`**

Think of this as a fork in the road. Python will read the boolean at the top, then if it is **`True`**, it will execute **`code_1`** not **`code_2`**

If it is **`False`**, it will skip **`code_1`** and only execute **`code_2`**

Read more about control flow on [Real Python](https://realpython.com/python-control-flow/).

````python
age = 19

if age >= 20:
    print("Come on in")
else:
    print("Go home kid")
````

For multiple conditions, one can use `if-elif-else`.

```python
    if condition_1:
        code_1
    elif condition_2:
        code_2
    else:
        code_3
```

````python
score = 85

if score >= 90:         
    grade = 'A'
elif score >= 80:
    grade = 'B'
else:
    grade = 'C'

print(f"Score: {score} → Grade: {grade}") # 
````

### 7. Loops

- **`for` loop**: iterates over an iterable (list, string, `range()`, etc.).

```python
for item in iterable:
    statement
```

- **`while` loop**: repeats while a condition is `True`. 

```python
while expression:
    statement
```

Read more on [W3Schools](https://www.w3schools.com/python/python_for_loops.asp).

````python
# for loop over a range
for i in range(4):       # 0, 1, 2, 3
    print(i, end=" ")
print()

# for loop over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# enumerate() — index + value together
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")

# while loop
count = 5
while count > 0:        # as long as count is larger than 0
    count -= 1          # reduce 1
    print(count)
````

### 8. Functions

Functions are defined with `def`. They can have default argument values, return multiple values (as a tuple), and include a docstring explaining their purpose.

**Scope:** variables created inside a function are **local** — they only exist within that function and are gone when it returns.

Read more on [W3Schools](https://www.w3schools.com/python/python_functions.asp).

````python
def greet(name, greeting="Hello"):
    """Return a greeting string. 'greeting' has a default value."""
    return f"{greeting}, {name}!"

print(greet("Alice"))                # uses default
print(greet("Bob", greeting="Hi"))   # override default

# Returning multiple values
def min_max(values):
    return min(values), max(values)

lo, hi = min_max([3, 1, 4, 1, 5, 9])
print(f"Min: {lo}, Max: {hi}")

# Lambda — a compact anonymous function
double = lambda x: x * 2
print(double(7))   # 14
````

### 9. Methods

A **method** is a function that belongs to an object. Where a standalone function is called as `function(object)`, a method is called *on* the object using dot notation: `object.method()`. Methods automatically have access to the object they are called on.

Every data type in Python comes with its own set of built-in methods, such as:

- str.upper()
- str.split()
- list.append()
- list.sort()
- dict.keys()

Below are a few examples. 

The key distinction:
- `len(my_list)` — built-in **function**, takes the object as an argument
- `my_list.append(4)` — **method**, called on the object itself

You will encounter this pattern constantly in Lecture 2 (`str.upper()`,`str.split()`,`list.append()`, `list.sort()`) and later throughout pandas (`df.head()`, `df.groupby()`, `series.mean()`, etc.).

Read more on [W3Schools](https://www.w3schools.com/python/python_lists_methods.asp).

````python
# String methods
text = "  hello, world  "
print(text.strip())           # remove whitespace: 'hello, world'
print(text.upper())           # 'HELLO, WORLD'
print(text.replace(",", ";")) # 'hello; world'

# List methods
numbers = [3, 1, 4, 1, 5]
numbers.append(9)             # modifies the list in place
numbers.sort()                # modifies the list in place
print(numbers)
````

### 10. Comments

- **Single-line comment**: starts with `#`, extends to end of line.
- **Multi-line comment**: use triple quotes `'''` or `"""`. Technically these are multi-line strings, often used as docstrings or block comments.

````python
'''
This is a multi-line comment, often used as a module or function docstring
to describe its purpose and usage.
'''

is_authenticated = True
has_permissions = True

if is_authenticated:
    # This block belongs to the if statement
    print("Authentication successful.")
    if has_permissions:
        # This is a nested if block
        print("Access granted.")
````

### 11. Indentation & Code Blocks

Python uses indentation to show **code hierarchy** — and it's not just formatting, it's logic and a clear visual structure. All statements in the same code block must use the same indentation level — the standard is **4 spaces**.

Here is an example:

```python
if score > 50:
    print("Pass")       # only runs if score > 50
    if score > 90:
        print("Distinction")  # only runs if score > 90
print("Done")           # always runs
```

- Each indentation level = one more layer of nesting
- Reading tip: follow the indentation to understand *what controls what*

````python
# Correct indentation
if True:
    print("True")
else:
    print("False")

# ❌ Inconsistent indentation causes an IndentationError:
# if True:
#     print("Answer")
#     print("True")
# else:
#     print("Answer")
#   print("False")    # Wrong indentation — runtime error
````

####  `None` & `NoneType`

`None` represents the **absence of a value**. It is its own type (`NoneType`) and is Python's equivalent of a null. A few important rules:

- Always check for `None` using `is None`, not `== None`
- `None` is not `0`, not `False`, not an empty string — it is uniquely itself
- A function that reaches its end without a `return` statement implicitly returns `None`

Read more on [W3Schools](https://www.w3schools.com/python/python_none.asp).

````python
result = None

print(result is None)     # True  — correct way to check
print(result == False)    # False — None is not False
print(result == 0)        # False — None is not zero
print(type(result))       # <class 'NoneType'>
````

### Execution Order and State

Above we looked at programming concepts with a language analogy, I hope that's helpful. 

However, in code, things don't remain static. Rather, they *change*. We are now looking at Execution Order and State.

Let's use a few examples to look, slowly and closely, at how *data* change while the code runs.

**Execution order** means code runs line by line, top to bottom (unless a loop or function call redirects it). The sequence matters:

```python
x = 5
x = x + 1
print(x)   # prints 6, not 5
```

**State** is the current value of all variables at any given moment during execution. As the program runs, state changes constantly.

```python
score = 0          # state: score is 0
score = score + 10 # state: score is now 10
score = score + 5  # state: score is now 15
print(score)       # prints 15
```

#### Why This Matters

- Reading code as static text will mislead you
- You need to **mentally trace execution** — imagine running the code step by step in your head
- Ask yourself at each line: *"What are the current values of my variables right now?"*

#### A Practical Technique: The State Table

When reading unfamiliar code, track state manually:

```python
a = 2
b = 3
a = a + b
b = a - b
a = a - b
```

| Line | a | b |
|---|---|---|
| `a = 2` | 2 | — |
| `b = 3` | 2 | 3 |
| `a = a + b` | 5 | 3 |
| `b = a - b` | 5 | 2 |
| `a = a - b` | 3 | 2 |

This is how experienced programmers debug code mentally. It's a skill, and it's learnable.

If you find the above State Table exercise useful, there is an exercise notebook named **Execution and State Tracing**. Work on it alongside this week's lectures. It may help to practice.

### A Few More Words On Reading Code

Above we have gone through some basic concepts and building blocks with simple code examples. 

Next we will have a look at a few tips to read Python code clearly. This helps us to develop good habit and learn to write, hopefully, well-structured clean code in the nearest future. 

#### Read Code Structurally

Don't start with details. Read in layers:

1. **Scan the structure** — where are the functions, loops, and conditionals?
2. **Follow the data flow** — what goes in, what comes out, what changes?
3. **Trace execution order** — what runs first? What triggers what?
4. **Read the details** — now zoom into individual lines

Think of it like reading a building blueprint:
- First: how many floors, where are the walls?
- Then: where do the pipes run?
- Finally: what are the light fittings?

#### A Path to Fluency

Learning to read code is like learning a language — with one key difference:

**Language fluency:** exposure + pattern recognition  
**Code fluency:** exposure + pattern recognition + **active execution tracing**

Progress looks like this:

- *Stage 1:* Can't understand individual lines
- *Stage 2:* Understand individual lines, but not the whole
- *Stage 3:* Understand blocks and structure
- *Stage 4:* Can predict what comes next
- *Stage 5:* Can spot bugs before running the code

The gap between Stage 2 and Stage 3 is usually structural reading. The gap between Stage 3 and Stage 4 is execution tracing.

### Practice the Four Building Blocks

These four things cover 80% of beginner code:

- **Variables** — how data is stored and named
- **Functions** — how actions are defined and called
- **Conditionals** — how decisions are made
- **Loops** — how repetition is controlled

Master these, plus the habit of **tracing execution and tracking state**, and you can read most entry-level project code.

Frameworks, classes, decorators, modules — those are later layers. Learn to read everyday conversation before tackling literature.

## Data Types

These four data types are also called collection data types or arrays.

### Tuple
An **ordered, immutable** sequence. Stored in round brackets as such `()`.

Use tuples for data that must not change — configuration values, coordinate pairs, fixed column schemas. Functions also return multiple values as tuples.

Read more on [W3Schools](https://www.w3schools.com/python/python_tuples.asp).

````python
# Tuple
dimensions = (1920, 1080)
x, y = dimensions               # unpacking
print(x, y)

# Try below, you will receive TypeError because tuples are immutable
# dimensions[0] = 800           
````

### Lists

A list is an **ordered, mutable** sequence. It can hold any mix of types. Lists are **the most common data container** you will work with before loading data into a DataFrame.

List uses square brackets `[]` and it can be created empty as a placeholder.

Below we'll have a look at indexing & slicing using `[]` and `:` and some common list methods.

#### Indexing and Slicing

Python index starts from 0. You can read about list index() method at [DataCamp](https://www.datacamp.com/tutorial/python-list-index).

````python
scores = [88, 72, 95, 60, 84, 91]

# Indexing and slicing
print(scores[0])       # first element: 88
print(scores[-1])      # last element: 91
print(scores[1:4])     # slice: [72, 95, 60]
print(scores[:3])      # first 3: [88, 72, 95]
print(scores[::2])     # every 2nd: [88, 95, 84]

# show all scores
for s in scores:
    print(s)

# show scores indexes
for s in scores:
    print(scores.index(s))

# or use enumerate() to show indexes next to list
for i, score in enumerate(scores):
    print(i, score) 
````

#### List Methods

> A **method** is a function that belongs to an object. Where a standalone function is called as `function(object)`, a method is called *on* the object using dot notation: `object.method()`. Methods automatically have access to the object they are called on.

When our object example is `scores`, the list method would look like `scores.method()`.

````python
# Modifying
scores.append(78)           # add to end
scores.insert(0, 100)       # insert at position. Here at index 0 (the first position)
scores.remove(60)           # remove first occurrence of value
popped = scores.pop()       # remove and return last element

# Useful list methods
print(len(scores))          # length
print(scores.count(88))     # how many times 88 appears
print(sorted(scores))       # returns a new sorted list
scores.sort(reverse=True)   # sorts in place
print(scores)
````

### Dictionaries

A dictionary stores **key-value pairs** (one pair or many pairs). Since Python 3.7 dictionaries preserve insertion order - that is *append at the end*. 

**Keys must be unique and immutable** (strings and numbers are the most common keys). In a certain sense, we can say *keys* are similar to *column names*.

In Pythonic-flavored data work, a *dict* often represents a single record —  that is one row of named fields. In such a sense, a *list* of *dict* is more comparable to a table that we are more familar with. 

A Dict is stored in curly braces as such `{}`.

Read more on [W3Schools](https://www.w3schools.com/python/python_dictionaries.asp).

````python
# an example of 1 dict

record = {
    "id": 108,
    "product": "Headset",
    "revenue": 5000,
    "in_store": True
}

# Access
print(record["product"])                  # KeyError if missing
print(record.get("margin", 0))            # returns 0 if missing

# Add / update
record["margin"] = 0.35
record["revenue"] = 5500                   # update existing key

# Remove
record.pop("in_store")                     # removes and returns the value

# Iterating — the three common patterns
for key in record:                        # iterate over keys
    print(key)

for value in record.values():             # iterate over values
    print(value)

for key, value in record.items():        # iterate over both
    print(f"{key}: {value}")

# Check if exists
print("revenue" in record)               # True
print("discount" in record)              # False
````

### Set

An **unordered collection of <font color=MediumOrchid>unique</font> elements**. Sets are optimised for membership testing and removing duplicates. This make it very handy in data analysis.

> So what type of brackets store a set?

That will be `{}`. Good to know that set() always returns a set object — it never preserves the original type, meaning if we apply set() on a list, the returned results will be a set, not a list. 

You can read more about **set** at [W3schools](https://www.w3schools.com/python/python_sets.asp).

````python
# Set
departments_raw = ["Sales", "HR", "Sales", "Finance", "HR", "HR"]
departments = set(departments_raw)          # duplicates removed
print(departments)                          # order is not guaranteed

# Fast membership test — O(1) regardless of set size
print("Sales" in departments)   # True

# Set operations — useful for comparing datasets
team_a = {"Alice", "Bob", "Carol"}
team_b = {"Bob", "Carol", "Dave"}

print(team_a & team_b)          # intersection: {'Bob', 'Carol'}
print(team_a | team_b)          # union: all members
print(team_a - team_b)          # difference: only in team_a
````

## 1. `if / elif / else`

Python checks conditions top to bottom and runs the **first** block that is `True`. 

```python
if condition:
    # runs if condition is True
elif other_condition:
    # runs if the first was False, this is True
else:
    # runs if none of the above were True
```


````python
temperature = 18

if temperature >= 25:
    print("Warm — maybe shorts weather.")
elif temperature >= 10:
    print("Mild — a light jacket should do.")
else:
    print("Cold — dress up properly.")
````

````python
# You can chain as many elif branches as you need
day = "Monday"

if day == "Monday":
    print("Start of the week.")
elif day == "Wednesday":
    print("Halfway there.")
elif day == "Friday":
    print("Weekend soon!")
else:
    print(f"{day} — just another day.")
````

## 5. `break` and `continue`

- **`break`** — exit the loop immediately
- **`continue`** — skip the rest of this iteration and go to the next one

```python
for item in sequence:
    if condition:
        break      # exit the loop entirely
    if other_condition:
        continue   # skip to the next iteration
```

````python
# break: stop as soon as we find what we're looking for
names = ["Alice", "Bob", "Charlie", "Diana", "Eve"]

for name in names:
    if name == "Charlie":
        print(f"Found Charlie! Stopping.")
        break
    print(f"Not Charlie: {name}")
````

````python
# continue: skip missing values, process the rest
values = [10, None, 5, None, 8, 3]

total = 0
skipped = 0

for v in values:
    if v is None:
        skipped += 1
        continue
    total += v

print(f"Total: {total}  |  Skipped: {skipped} missing values")
````

## A First Look at Data Workflow

With these basics (before pandas and other packages), you can already do meaningful data work using just lists of dicts, comprehensions, and built-ins. This is also the mental model pandas builds on.

Here is a small end-to-end example: load, clean, filter, aggregate, and report.

💡 Before we formally work through Functions ``def`, can you see what problem below data workflow might have?

````python
# --- Raw data (as you might receive from an API or CSV) ---
raw_data = [
    {"region": "North", "product": "Widget A", "revenue": "120000", "cost": "80000"},
    {"region": "South", "product": "Widget B", "revenue": "85000",  "cost": "60000"},
    {"region": "North", "product": "Widget C", "revenue": None,     "cost": "40000"},
    {"region": "East",  "product": "Widget A", "revenue": "200000", "cost": "130000"},
    {"region": "South", "product": "Widget C", "revenue": "30000",  "cost": "20000"},
]

# --- Step 1: Parse & clean ---
def parse_record(r):
    try:
        rev = float(r["revenue"])     # convert string to number
        cost = float(r["cost"])
        return {
            "region":  r["region"],
            "product": r["product"],
            "revenue": rev,
            "cost":    cost,
            "profit":  rev - cost,
            "margin":  (rev - cost) / rev
        }
    except (TypeError, ValueError):
        return None   # skip rows with missing/bad data

clean = [parse_record(r) for r in raw_data]
clean = [r for r in clean if r is not None]   # drop failed rows

print(f"Rows after cleaning: {len(clean)}")

# --- Step 2: Filter ---
high_margin = [r for r in clean if r["margin"] >= 0.30]
print(f"High-margin rows: {len(high_margin)}")

# --- Step 3: Aggregate by region ---
regions = {r["region"] for r in clean}

for region in sorted(regions):
    rows = [r for r in clean if r["region"] == region]
    total_rev = sum(r["revenue"] for r in rows)
    total_profit = sum(r["profit"] for r in rows)
    print(f"{region:8s}  Revenue: {total_rev:>10,.0f}  Profit: {total_profit:>10,.0f}")

# --- Step 4: Top product by revenue ---
top = sorted(clean, key=lambda r: r["revenue"], reverse=True)[0]
print(f"\nTop product: {top['product']} in {top['region']} — {top['revenue']:,.0f}")
````

### Mutable vs Immutable Objects

This is one of the most common sources of unexpected bugs for newcomers. Understanding it now saves a lot of debugging later.

**Mutable** objects (`list`, `dict`, `set`) can be changed after creation.  
**Immutable** objects (`int`, `float`, `str`, `tuple`) cannot.

The subtle problem: assigning a mutable object to a new variable does **not** copy it. Both variables point to the **same object** in memory.

````python
# --- The problem ---
original = [10, 20, 30]
alias = original        # NOT a copy — both point to the same list

alias.append(99)
print(original)         # [10, 20, 30, 99] — original was changed too!

# --- The fix: copy explicitly ---
original = [10, 20, 30]
copy = original.copy()  # or: list(original)  or  original[:]

copy.append(99)
print(original)         # [10, 20, 30]  — untouched
print(copy)             # [10, 20, 30, 99]

# Same issue with dicts
rec = {"name": "Alice", "score": 88}
rec_copy = rec.copy()
rec_copy["score"] = 0
print(rec)              # original unchanged

# Immutable: no such issue
a = "hello"
b = a
b = b.upper()           # creates a NEW string, does not modify a
print(a)                # 'hello' — unchanged
````

### `*args` and `**kwargs`

You will encounter these in library documentation and in functions designed to accept a flexible number of arguments.

- `*args` — collects any number of **positional** arguments into a **tuple**
- `**kwargs` — collects any number of **keyword** arguments into a **dict**

The names `args` and `kwargs` are conventions — the `*` and `**` are what matter.

````python
# *args — variable number of positional arguments
def total(*amounts):
    return sum(amounts)

print(total(100, 200))           # 300
print(total(10, 20, 30, 40))     # 100

# **kwargs — variable number of keyword arguments
def describe_record(**fields):
    for key, value in fields.items():
        print(f"  {key}: {value}")

describe_record(name="Widget A", revenue=50_000, region="North")

# In practice: you'll see these when calling pandas, matplotlib, etc.
# e.g. df.plot(kind="bar", title="Revenue", figsize=(10,5), color="steelblue")
# All those extra keyword arguments go into **kwargs inside the function.
````

## Function `print_tree()`

Here we will start to transition using Python libraries and packages. To run this function, we will use a standard Python library `pathlib`. It comes with Python and no additional installation needed.

To read more about `pathlib`, go to [W3Schools](https://www.w3schools.com/python/ref_module_pathlib.asp) for the shortest version. Or visit [Real Python](https://realpython.com/python-pathlib/) for more detailed info.

````python
from pathlib import Path

def print_tree(path, prefix=""):
    '''
    print a given folder's entire structure and underlyting files. 
    '''
    contents = sorted(Path(path).iterdir())
    for i, item in enumerate(contents):
        connector = "└── " if i == len(contents) - 1 else "├── "
        print(prefix + connector + item.name)
        if item.is_dir():
            extension = "    " if i == len(contents) - 1 else "│   "
            print_tree(item, prefix + extension)
````

## 1. What is a Function?

A function is a **named, reusable block of code**. You define it once, and call it as many times as you like.

**Syntax:**
```python
def function_name(arguments):
    # code block
    return result
```

- `def` — tells Python you're defining a function
- `function_name` — you choose this name (use lowercase with underscores, aka *snake_case*)
- `arguments` — fancy name for inputs the function receives. It's optional.
- `return` — optional output the function sends back

Think of a function like a **recipe** as in cooking recipe: it has a name, takes some ingredients (arguments), does something, and optionally produces something (a return value).

````python
# The simplest possible function — no arguments, no return value
def say_hello():
    print("Hello!")

say_hello()  # Call the function
````

````python
# A function with an argument
def greet(name):
    print("Hello,", name)

greet("Pia")
greet("Fabian")
````

> **Note:** Defining a function doesn't run it. You must **call** it with `function_name()` to execute the code inside.

## 2. Returning Values

So far our functions just print things. 

In real data work, you usually want a function to **give you back a value** you can use elsewhere. The **returned value** is often an object like a cleaned list, a dictionary of results, or later on, a full DataFrame.

That's what `return` does — it sends a value back to whoever called the function.

````python
import sys
print(sys.executable)
````

````python
# In below example, we only print without return — result disappears after print
def add_no_return(a, b):
    print(a + b)  # this just prints, doesn't store the result

result = add_no_return(3, 4)  # result is None!
print("result is:", result)
````

````python
# Here we use return — we get the value returned every time we call the function with valid input
def add(a, b):
    return a + b

result = add(3, 4)  # result stores 7
print("result is:", result)


# Now we can use it!
print(result * 2)
print(add(10, add(3, 4)))  # functions inside functions

# This particular function works with any "addable" elements
result = add("Hello", "!")
print(result)

result = add("Hello", "!"*3)
print(result)

# Uncomment to run below and observe the TypeError message
result = add({"name":"Pia"}, {"name":"Fabian"})
print(result)
````

````python
# Returning multiple values — Python packs them into a tuple
def min_max(numbers):
    return min(numbers), max(numbers)

low, high = min_max([4, 1, 9, 2, 7])
print("Min:", low)
print("Max:", high)

# Note again that unless the function defines input type, this function works also for other non-numericals
start, end = min_max(["k", "d", "y", "i", "o"])
print("Start:", start)
print("End:", end)

# Define specific numerical types
def min_max_num(numbers:int):
    return min(numbers), max(numbers)

low_int, high_int = min_max_num([4, 1, 9, 2, 7])
print("Min number:", low_int)
print("Max number:", high_int)

# What does you think below would return?
low_float, high_float = min_max_num([4.6, 1.2, 9.7, 2, 7])
print("Min number:", low_float)
print("Max number:", high_float)
print(type(low_float))
````

> **Rule of thumb:** If you'll need the result later (store it, pass it on, compare it), use `return`. If you just want to display something right now, use `print`.

## 3. `=` & Optional Argument's Default Value

To explain how this work, let's use earlier functions as examples. 

Below partially-shown functions require arguments as such. All must be entered, otherwise there will be errors.

```python
def add(a, b)
#       -  -
#       required, required

def min_max(numbers)
#           -------
#           required
```

Compare to below modifications:

```python
def add(a, b, unit_of_measure="piece/s")
#       -  -  ^^^^^^^^^^^^^^^^^^^^^^^^
#       required, required, optional with default value (use `=` to define)
```

````python
def add(a:int, b:int, unit_of_measure="pieces"):
        return(f"{a + b} {unit_of_measure}")

result = add(3, 4)  # result takes default value for u_o_m
print(f"We have in total {result} in stock from both warehouses.")

result = add(2, 3, "boxes")  # result takes input value instead for u_o_m
print(f"We have in total {result} of tomato sauce counting both in the kitchen and the summer house.")
````

To summarize: you can give function's optional arguments **default values**. If the caller doesn't provide that argument (input), the default value is used. This makes functions more flexible while explicit at the same time.

Here the **sequence of arguments defined** is very imporant:

- Defaults (optional arguments) come after required arguments in the function definition.
- Default values for optional arguments must use `=` to indicate they are optional.

In addition:

- You can use `optional_argument_1=None`: used when the feature is off by default.
- You can use `optional_argument_1=True / False`: used it's a toggle with a natural default state. 

````python
# currency is optional — defaults to "SEK" if not provided
def format_price(amount, currency="SEK"):
    return f"{amount:.2f} {currency}"

print(format_price(1500))          # uses default
print(format_price(1500, "EUR"))   # override the default
print(format_price(1500, "USD"))   # another override
````

````python
# Real-world example: cleaning a text column
# Default behavior is to strip whitespace and lowercase — but both are optional
def clean_text(text, strip=True, lowercase=True):
    if strip:
        text = text.strip()
    if lowercase:
        text = text.lower()
    return text

print(clean_text("  Stockholm  "))           # both defaults apply
print(clean_text("  Stockholm  ", lowercase=False))  # keep original case
print(clean_text("  Stockholm  ", strip=False))       # keep spaces
````

> **A good practice:** pass arguments (*in plain english: enter input for arguments*) by variable **names** — this makes code easier to read and avoids mistakes with argument order because Python reads the **positions of arguments** not the content. See below examples:

````python
def register_student(name, age, course, active=True):
    print(f"{name}, age {age}, enrolled in {course}, active: {active}")

# ❌ Positional entry — this won't throw an error technically, but it is wrong
print("\n------ wrong data entered at wrong position ------")
register_student("BI & Python", 28, "Pia" )

print("\n------ corret & easy to read ------")
# ✅ Keyword — order doesn't matter, meaning is clear
register_student(name="Pia", age=28, course="BI & Python")

# ✅ Also fine — mix positional for obvious args, keyword (variable name) for the rest
register_student("Pia", 28, course="BI & Python", active=False)
````

## 4. Flexible Arguments: `*args` and `**kwargs`

Sometimes you don't know in advance how many arguments will be passed. Python has two special syntaxes for this:

- `*args` — collects any number of **positional** arguments into a tuple
- `**kwargs` — collects any number of **keyword** arguments into a dictionary

The names `args` and `kwargs` are just conventions — the `*` and `**` are what matter.

> **A heads-up:** It's fine to just know about `*args` and `**kwargs` for now. Especially `**kwargs` is more advanced in function writing. Let's look into this later in our course when we have actual examples suitable to explain and demostrate how to use.

````python
# *args — accept any number of values
def total(*args):
    print("Received:", args)  # args is a tuple
    return sum(args)

print(total(10, 20))
print(total(10, 20, 30, 40, 50))
````

````python
# **kwargs — accept any number of named values
def show_info(**kwargs):
    print("Received:", kwargs)  # kwargs is a dictionary
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

show_info(name="Pia", department="Finance", city="Stockholm")
````

## 5. Namespaces and Scope

This is one of those topics that may be confusing at beginning but is very important once you write bigger programs.

**Scope** = where a variable "lives" and can be accessed.

- Variables created **inside** a function are **local** — they only exist inside that function.
- Variables created **outside** functions are **global** — they exist everywhere.

Python always looks for a variable in the **local scope first**, then the global scope.

````python
# x inside the function is LOCAL — a different x than outside
x = 100  # global x

def change_x():
    x = 999  # local x — only lives inside this function
    print("Inside function, x =", x)

change_x()
print("Outside function, x =", x)  # still 100 — global was NOT changed
````

````python
# The function CAN read a global variable (but shouldn't change it this way)
company_name = "Acme AB"

def get_greeting():
    return f"Welcome to {company_name}"  # reads the global

print(get_greeting())
````

````python
# Accessing a local variable OUTSIDE its function → NameError
def make_greeting():
    greeting = "Hello!"  # local variable.
    return greeting

make_greeting()
print(greeting)  # ← uncomment this to see NameError
````

> **Key insight:** Functions should work with their **inputs and outputs**, not rely on or change global state. This makes them predictable and reusable.

## 6. Anonymous Functions: `lambda`

A `lambda` is a **small, one-line function without a name**. It's useful when you need a quick function to pass into another function — like when sorting or filtering data.

**Syntax:**
```python
lambda arguments: expression
```

A lambda always returns the result of the expression — no `return` keyword needed.

````python
# Regular function vs lambda — same result, different style
def double(x):
    return x * 2

double_lambda = lambda x: x * 2

print(double(5))
print(double_lambda(5))
````

````python
# A very common use: sorting by a custom key
products = [
    {"name": "Laptop",  "price": 12000},
    {"name": "Mouse",   "price": 299},
    {"name": "Monitor", "price": 4500},
]

# Sort by price using a lambda
sorted_products = sorted(products, key=lambda p: p["price"])
sorted_products_desc = sorted(products, key=lambda p: p["price"], reverse=True)

print("--- sorted in ascending order ---")
for p in sorted_products:
    print(p["name"], "—", p["price"])

print("\n--- sorted in descending order ---")
for p in sorted_products_desc:
    print(p["name"], "—", p["price"])
````

````python
# Another common use: filtering with a lambda
prices = [299, 12000, 4500, 89, 7800, 150]

expensive = list(filter(lambda p: p > 1000, prices)) 
# above can also be achieved with a list comprehension, such as expensive = [p for p in prices if p > 1000]
print("Expensive items:", expensive)
````

````python
# Filter restaurants by minimum rating — lambda as the condition
restaurants = [
    {"name": "Pasta Palace", "rating": 4.5},
    {"name": "Burger Barn",  "rating": 3.8},
    {"name": "Sushi Spot",   "rating": 4.9},
]

top_rated = [r for r in restaurants if r["rating"] >= 4.0]
top_rated_names = list(map(lambda r: r["name"], top_rated))
print("Top rated:", top_rated_names)
````

> **When to use lambda vs `def`?**
> - Use `lambda` for quick, throwaway functions — especially as arguments to `sorted()`, `filter()`
> - Use `def` for anything you'll name, reuse, or that has more than one line of logic
>
> In pandas, `lambda` can be used conveniently so — e.g. `df["col"].apply(lambda x: x * 1.25)`

## 7. Error Handling: `try` / `except`

When working with real data, errors happen — missing values, wrong data types, files that don't exist. Instead of letting your script crash, you can **catch errors** and handle them gracefully.

**Syntax:**
```python
try:
    # code that might fail
except SomeError:
    # what to do if it fails
```

````python
# Without error handling — this crashes the whole script
def divide(a, b):
    return a / b

# divide(10, 0)  # ← uncomment to see ZeroDivisionError
````

````python
# With error handling — we catch the error and decide what to do
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Warning: cannot divide by zero, returning None")
        return None

print(safe_divide(10, 2))
print(safe_divide(10, 0))
````

````python
# Catching multiple error types
def parse_number(value):
    try:
        return float(value)
    except ValueError:
        print(f"Could not convert '{value}' to a number")
        return None
    except TypeError:
        print(f"Wrong type: {type(value)}")
        return None

print(parse_number("3.14"))   # works
print(parse_number("hello"))  # ValueError
print(parse_number(None))     # TypeError
````

````python
# else and finally — advanced but useful
# else: runs only if NO error occurred
# finally: ALWAYS runs, regardless of error

def load_value(data, key):
    try:
        result = data[key]
    except KeyError:
        print(f"Key '{key}' not found")
        result = None
    else:
        print(f"Successfully retrieved: {result}")
    finally:
        print("Done.")  # always runs
    return result

sales = {"Q1": 150000, "Q2": 210000}
load_value(sales, "Q1")
print() # print() with no arguments simply print an empty line - a formatting trick for better readability.
load_value(sales, "Q3")
````

### Errors

| Error | When it happens | Example |
|---|---|---|
| `ValueError` | Wrong value for the type | `float("N/A")` |
| `TypeError` | Wrong type entirely | `float(None)` |
| `KeyError` | Dict key doesn't exist | `row["revenue"]` when key is missing |
| `IndexError` | List index out of range | `my_list[99]` on a short list |
| `AttributeError` | Method doesn't exist on that type | `"hello".upper` without `()` |
| `FileNotFoundError` | File path is wrong or missing | `pd.read_csv("wrong_path.csv")` |
| `ZeroDivisionError` | Dividing by zero | `revenue / covers` when covers is 0 |

<br>

> You don't need to memorise these. Python always tells you the error type in the traceback — the skill is **recognising** them when they appear, not recalling them from memory.

## Putting It All Together: A Mini Data Pipeline

Here's a small example combining functions, defaults, lambda, and error handling — the kind of code you may actually write in data projects.

````python
# Raw sales data — messy, as usual in real life
raw_data = [
    {"product": "  Laptop ? ", "revenue": "120000"},
    {"product": "MoUSe",      "revenue": "N/A"},
    {"product": "MONITOR#",    "revenue": "45000"},
    {"product": "keyboard! ",  "revenue": "8500"},
]

# Step 1: Clean product names
# use built-in regex package to clean
import re 

def clean_name(name):
    name=re.sub("[!#?]", "", name)
    # below shows how the functions are chained up with . instead of having two lines of code
    return name.strip().title()

# Step 2: Safely parse revenue
# Below default = 0 is a more elegant way than the if/else we used earlier
def clean_revenue(value, default=0):
    try:
        return float(value)
    except ValueError:
        print(f"  Could not parse revenue '{value}', using {default}")
        return default

# Step 3: Process the dataset
def process_sales(data):
    cleaned = []
    for row in data:
        cleaned.append({
            "product": clean_name(row["product"]),
            "revenue": clean_revenue(row["revenue"])
        })
    return cleaned

# Step 4: Sort by revenue using lambda
result = process_sales(raw_data)
result_sorted = sorted(result, key=lambda r: r["revenue"], reverse=True)

print("\nFinal cleaned & sorted data:")
for row in result_sorted:
    print(f"  {row['product']:12} — {row['revenue']:>10,.0f} SEK")
````

### pip/shell vs. help()

**The difference below the two approaches are**:

- `%pip show pandas` is a **pip/shell command** that shows package metadata — version, author, install location, dependencies, etc. 
- `help('pandas')` uses **Python's built-in help system** and shows the module's docstring and public API — classes, functions, and their documentation.

More specifically:

- `%pip show pandas` reads from the package's dist-info **metadata** folder (e.g. `.venv/lib/pandas-2.x.x.dist-info/`). This is pure packaging metadata written by pip at install time — it knows nothing about the code itself.
- `help('pandas')` reads from the actual **source code** inside `.venv/lib/pandas/` — specifically the docstrings embedded in the .py files.

Now open your `.venv` folder and look for the following.

```bash
.venv/lib/
├── pandas/                  ← help() reads from here (source + docstrings)
│   ├── __init__.py
│   ├── core/
│   └── ...
└── pandas-2.x.x.dist-info/ ← pip show reads from here (install metadata)
    ├── METADATA
    ├── RECORD
    └── ...
```

> Now you know **"where does pandas live?" in this `.venv`** on your laptop.

| Method | Selects by | Example |
|--------|-----------|----------|

### 5. Cleaning data — the pandas way

In earlier lectures, we did quite a few *manual* cleaning by writing functions and loops from Python built-ins. Pandas has built-in tools for most of these jobs.

Let's reload the *raw* data and clean it with pandas this time.

````python
# Raw data — same as Lecture 4
raw_data = [
    {"restaurant": "  pasta palace   ",  "city": "Stockholm",  "revenue": "182000",    "currency": "SEK", "covers": "210", "rating": "4.5"},
    {"restaurant": "Burger Barn  ",   "city": "Stockholm",  "revenue": "95000",     "currency": "SEK", "covers": "180", "rating": "3.8"},
    {"restaurant": "SUSHI SPOT",      "city": "Copenhagen", "revenue": "28000",     "currency": "DKK", "covers": "95",  "rating": "4.9"},
    {"restaurant": "taco town",       "city": " Oslo ",       "revenue": "61000",     "currency": "NOK", "covers": "140", "rating": None},
    {"restaurant": "Pasta Palace",    "city": "Stockholm",  "revenue": "182000",    "currency": "SEK", "covers": "210", "rating": "4.5"},
    {"restaurant": "Green Garden",    "city": "Oslo",       "revenue": "see notes", "currency": "NOK", "covers": "88",  "rating": "4.1  "},
    {"restaurant": "Burger Barn",     "city": "Copenhagen", "revenue": "31500",     "currency": "DKK", "covers": "210", "rating": "6.0"},
    {"restaurant": "  Ramen House ",  "city": "Stockholm",  "revenue": "74000",     "currency": "SEK", "covers": None,  "rating": "4.2"},
]

raw = pd.DataFrame(raw_data)
raw
````

#### 5a. Fix restaurant names

In pandas: `Series.str` — a string accessor that applies string methods to every value at once.

````python
# .str.strip() removes leading/trailing whitespace from every value
# .str.title() applies title case to every value
# Chain them together — just like chaining string methods in Python

raw["restaurant"] = raw["restaurant"].str.strip().str.title()
raw["restaurant"]

# Compare to function written with built-ins:
# def clean_name(name):
#     return name.strip().title()
# [clean_name(r["restaurant"]) for r in raw_data]
````

#### About method chaining

The above method chaining is very useful and concise. However, when it becomes too long, it will be hard to read. 

Therefore, below code formatting makes the chaining easier to read. But it is important to test your code so the breaks are correct wrapped without breaking code.

````python
# An alternative way to chain methods

raw["restaurant"] = (raw["restaurant"]
                    #comment
                    .str.strip()
                    #comment
                    .str.capitalize())

raw["restaurant"]
````

#### 5b. Convert revenue to numeric

In earlier lecture, we used: `parse_number()` — `try/except` around `float(value)`  
In pandas: `pd.to_numeric()` with `errors='coerce'` — bad values become `NaN` automatically.

````python
# check data types before cleansing

raw[["revenue", "covers", "rating"]].dtypes
````

````python
# errors='coerce' turns anything that can't be converted into NaN
# 'see notes' becomes NaN — no try/except needed

raw["revenue"] = pd.to_numeric(raw["revenue"], errors="coerce")
raw["covers"]  = pd.to_numeric(raw["covers"],  errors="coerce")
raw["rating"]  = pd.to_numeric(raw["rating"],  errors="coerce")
````

````python
# After cleansing

raw[["revenue", "covers", "rating"]].dtypes
````

````python
raw[["revenue", "covers", "rating"]]
````

#### 5c. Handle missing values

In earlier lecture: `if value is None: return default`  
In pandas: `.fillna()` to replace NaN with a default, or `.dropna()` to remove rows with missing values.

````python
# Replace NaN revenue with 0 (same default we used in parse_number)
raw["revenue"] = raw["revenue"].fillna(0)

raw[["revenue", "rating"]]
````

````python
# For rating: cap values above 5.0 — then leave NaN as-is (no valid rating to fall back on)
raw["rating"] = raw["rating"].clip(upper=5.0)

# Observe row index 6
raw[["revenue", "rating"]]
````

#### 5d. Remove duplicates

In pandas: `.drop_duplicates()` — one line.

````python
# After cleansing, below rows appear to be identical

raw.iloc[[0,4]]
````

````python
# Keep first occurrence

raw = raw.drop_duplicates(subset=["restaurant", "city", "revenue", "currency", "covers","rating"], keep="first")
````

````python
# After dedup

raw
````

### 5. Cleansing data with `map` and `apply`

Compare below with lecture 5 same section. Do you see how they differ?

````python
# 5a. fix multiple string columns

raw_1[["restaurant", "city"]] = (raw_1[["restaurant", "city"]]
                                   .map(lambda x: x.strip().capitalize()))

# 5b & 5c. convert to numeric with apply and chain in fillna for missing data

raw_1[["revenue","covers"]] = (raw_1[["revenue","covers"]]
                               .apply(pd.to_numeric,errors="coerce")
                               .fillna(0))

# 5b & 5c. convert to numeric and handle irregular values
# Rating with NaN is probably meaningful for this business case. Here we choose to keep it as such instead of replacing with 0.

raw_1["rating"]  = (pd.to_numeric(raw_1["rating"],  errors="coerce")
                    .clip(upper=5.0))

# 5d: dedup when certain they are duplicates

raw_1 = raw_1.drop_duplicates(subset=["restaurant", "city"], keep="first")

raw_1
````

````python
raw_2 = raw_1.copy()
````

Next, we want to convert revenues to the same currency SEK. We do so by adding two columns by adding a calcuated SEK amount as a new column. 

````python
# For currency conversion we create a FX (foreign exchange) dictionary - a simple mapping table.

EXCHANGE_RATES = {"SEK": 1.0, "DKK": 1.45, "NOK": 0.98}

# new column with FX rate
raw_2["rate"] = raw_2["currency"].map(EXCHANGE_RATES)

# new calculated column for revenue in SEK
raw_2["revenue_sek"] = raw_2["revenue"] * raw_2["rate"]

raw_2
````

````python
# New column revenue per cover — pandas handles division column-by-column
# Where covers is NaN or 0, the result is NaN automatically (no ZeroDivisionError)

raw_2["rev_per_cover"] = (raw_2["revenue_sek"] / raw_2["covers"]).round(1)
````

For the next new column revenue_tier, we want to label rows (restaurants) based on the size of revenue. We will look at three alternative ways to get this done.

````python
# Option 1: the pandas native pandas.cut function

# first without label the range (interval)
raw_2["revenue_tier"] = pd.cut(raw_2["revenue_sek"],
                               bins=[0, 50000, 100000, float("inf")]
                               )
raw_2
````

````python
# clearly labeled

raw_2["revenue_tier"] = pd.cut(raw_2["revenue_sek"],
                               bins=[-float("inf"), 50000, 100000, float("inf")],
                               labels=["Low", "Medium", "High"])

raw_2
````

````python
# What is float("inf")? Try remove float("inf") or replace it with say 120000 and see what happens.

raw_2["revenue_tier"] = pd.cut(raw_2["revenue_sek"],
                               bins=[0, 50000, 100000, 120000],
                               labels=["Low", "Medium", "High"])

raw_2
````

````python
# Option 2: a custom function AND .apply()

def revenue_tier(rev):
    if rev >= 100000:
        return "High"
    elif rev >= 50000:
        return "Medium"
    else:
        return "Low"

raw_2["tier"] = raw_2["revenue_sek"].apply(revenue_tier)
````

````python
# Option 3: with a lambda and .apply. 

# below can be written in one line. I personally find it hard to read - but maybe it works after seeing it often:) I reformatted the code, slightly easier to read.

# raw_2["tier"] = raw_2["revenue_sek"].apply(lambda x:"High" if x >= 100000 else "Medium" if x >= 50000  else "Low")

raw_2["tier"] = (raw_2["revenue_sek"]
                 .apply(lambda x: 
                        "High" if x >= 100000 
                        else "Medium" if x >= 50000 
                        else "Low"))
````

# Python Regex for Beginners

A friendly guide based on the [official Python docs](https://docs.python.org/3/howto/regex.html)

## What is Regex?

A **regular expression** (regex) is a pattern you use to search for text. Think of it like a super-powered "Find" tool.

For example:
- Find all phone numbers in a document
- Check if an email address looks valid
- Pull all dates out of a log file

In Python, regex is available through the **built-in `re` module**. Run the cell below to import it — you'll need this in every session!

````python
import re  # The re module is built into Python — no install needed.
````

## Simple Patterns

The simplest regex is just the text you want to find. The pattern `cat` matches the word `cat` exactly.

`re.search()` returns a **match object** if found, or `None` if not. Try both cells below:

````python
# This should find a match
result = re.search("cat", "I have a cat")
print(result) 
````

````python
# ❌ This should return None (no match)
result = re.search("cat", "I have a dog")
print(result)
````

## Special characters (metacharacters)

Some characters have special meaning in regex — they are **not** matched literally:

. ^ $ * + ? { } [ ] \ | ( )

We'll cover the most important ones below.

## Raw string `r""`

Before we go further — this is the **#1 tip for beginners**.

Always prefix your regex pattern with `r`. This tells Python not to treat backslashes (`\`) specially, which avoids a lot of confusing bugs.

````python
# Always do this instead — the r prefix makes it a "raw string"
result = re.search(r"\d", "room 42")
print(result)

# ❌ Avoid this — can cause unexpected behaviour
re.search("\d", "room 42")
````

## Character Classes `[ ]`

Square brackets let you match **one character from a set**.

| Pattern | Matches |
|---------|--------|
| `[abc]` | Any one of: `a`, `b`, or `c` |
| `[a-z]` | Any lowercase letter |
| `[A-Z]` | Any uppercase letter |
| `[0-9]` | Any digit |
| `[^abc]` | Anything **except** `a`, `b`, or `c` |

<br>
The `^` inside brackets means "NOT".

````python
# Match any vowel
m = re.search(r"[aeiou]", "hello")
print(m.group())  # 'e' — the first vowel found
````

````python
# Match any digit
m = re.search(r"[0-9]", "room 42")
print(m.group())  # '4'
````

````python
# Try it yourself! Change the pattern or text and see what matches.
my_pattern = r"[A-Z]"
my_text = "Hello World"

m = re.search(my_pattern, my_text)
print(m.group() if m else "No match")
````

## Handy Shortcut Sequences

Instead of writing `[0-9]` every time, Python offers shortcuts:

| Shortcut | Meaning | Same as |
|----------|---------|--------|
| `\d` | Any digit | `[0-9]` |
| `\D` | Any non-digit | `[^0-9]` |
| `\w` | Any letter, digit, or underscore | `[a-zA-Z0-9_]` |
| `\W` | Any non-word character | `[^a-zA-Z0-9_]` |
| `\s` | Any whitespace (space, tab, newline) | `[ \t\n\r]` |
| `\S` | Any non-whitespace | `[^ \t\n\r]` |
| `.` | Any character except newline | — |

````python
print(re.search(r"\d", "I am 25 years old").group())  # '2'
print(re.search(r"\w", "Hello!").group())              # 'H'
print(re.search(r"\s", "Hello World").group())         # ' ' (a space)
````

## Repeating Things (Quantifiers)

What if you want to match more than one character?

| Symbol | Meaning | Example | Matches |
|--------|---------|---------|--------|
| `*` | 0 or more times | `ca*t` | `ct`, `cat`, `caaat` |
| `+` | 1 or more times | `ca+t` | `cat`, `caaat` (not `ct`) |
| `?` | 0 or 1 time (optional) | `colou?r` | `color`, `colour` |
| `{n}` | Exactly n times | `\d{4}` | `2024` |
| `{m,n}` | Between m and n times | `\d{2,4}` | `12`, `123`, `1234` |

````python
# * means zero or more — so 'ct' still matches ca*t
print(re.search(r"ca*t", "ct").group())     # 'ct'
print(re.search(r"ca*t", "caaat").group())  # 'caaat'
````

````python
# + means one or more — 'ct' does NOT match ca+t
print(re.search(r"ca+t", "ct"))      # None
print(re.search(r"ca+t", "cat").group())  # 'cat'
````

````python
# ? makes something optional
print(re.search(r"colou?r", "color").group())   # 'color'
print(re.search(r"colou?r", "colour").group())  # 'colour'
````

````python
# {n} matches exactly n times
print(re.search(r"\d{4}", "Year: 2024").group())  # '2024'
````

## The 4 Main Functions

### `re.search()` — Find the first match *anywhere* in the string

````python
text = "I have 2 cats and 3 dogs"

m = re.search(r"\d", text)
print(m.group())  # '2' — first digit found
````

### `re.match()` — Only checks the very *beginning* of the string

````python
print(re.match(r"\d", "I have 2 cats"))  # None — doesn't start with a digit
print(re.match(r"\d", "42 cats").group())  # '4' — starts with a digit ✅
````

### `re.findall()` — Returns a *list* of all matches

````python
text = "I have 2 cats and 3 dogs"

numbers = re.findall(r"\d", text)
print(numbers)  # ['2', '3']
````

### `re.sub()` — Find and *replace*

````python
text = "I have 2 cats and 3 dogs"

result = re.sub(r"\d", "X", text)
print(result)  # 'I have X cats and X dogs'
````

## Checking for a Match Safely

`search()` returns `None` if nothing is found. Calling `.group()` on `None` will crash your code! Always check first:

````python
m = re.search(r"\d+", "room 42")

if m:
    print("Found:", m.group())   # Found: 42
else:
    print("No match")
````

## Match Object Methods

When you get a match, you can ask it for details:

````python
m = re.search(r"\d+", "room 42")

print(m.group())   # '42'    — the matched text
print(m.start())   # 5       — where the match starts
print(m.end())     # 7       — where it ends
print(m.span())    # (5, 7)  — start and end as a tuple
````

## Useful Flags

Flags tweak how matching works. Pass them as a second argument:

| Flag | Short | What it does |
|------|-------|--------------|
| `re.IGNORECASE` | `re.I` | Case-insensitive matching |
| `re.MULTILINE` | `re.M` | `^` and `$` match each line, not just the whole string |
| `re.DOTALL` | `re.S` | `.` also matches newline characters |

````python
# Without flag — case sensitive, no match
print(re.search(r"hello", "HELLO WORLD"))  # None

# With IGNORECASE flag — matches!
print(re.search(r"hello", "HELLO WORLD", re.IGNORECASE).group())  # 'HELLO'
````

## Example: Find Email Addresses

````python
text = "Contact us at hello@example.com or support@company.org"

emails = re.findall(r"\w+@\w+\.\w+", text)
print(emails)

# ['hello@example.com', 'support@company.org']
````

## 1. Raw data ingestion (generation)

````python
raw_data = [
    {"region": "North", "product": "Widget A", "revenue": "120000", "cost": "80000"},
    {"region": "South", "product": "Widget B", "revenue": "85000",  "cost": "60000"},
    {"region": "North", "product": "Widget C", "revenue": None,     "cost": "40000"},
    {"region": "East",  "product": "Widget A", "revenue": "200000", "cost": "130000"},
    {"region": "South", "product": "Widget C", "revenue": "30000",  "cost": "20000"},
]
````

## 2. Inspect data

We will learn in the future lecture how to profile data with Pandas. For now, we eye ball it. Or you can also use VSCode extension Data Wrangler.

## 3. Create cleansing function

In comments, we decribe what the code does or the logic behind. Please note this is for learning purpose. At work, you ought to be concise and write *clean* code whenever possible.

````python
# This function will go through record row by row when called for.

# function requires a single argument 'r':
def parse_record(r): 
    # __docstring__ is added
    """
    Convert revenue and cost from string to number. 
    Calculate and add new key value pairs profit and margin.
    All record with None value are replaced with 0.
    Margin is treated as 0 when revenue is 0 (None).
    """
    # Use `try / except` to catch errors gracefully instead of letting the program crash. You can catch specific exception types for precise control.
    try:
        # use a ternary expression to insert 0 if field value is None
        rev = 0 if r["revenue"] is None else float(r["revenue"])
        cost = 0 if r["cost"] is None else float(r["cost"])
        # return dicts containing below key and value pairs
        return {
            "region":  r["region"],
            "product": r["product"],
            "revenue": rev,
            "cost":    cost,
            # new keys profit and margin are added
            "profit":  rev - cost,
            # if revenue is 0, margin would show ZeroDivisionError. We let the roll level margin be 0 for these.
            "margin":  0 if rev == 0 else round((rev - cost) / rev, 2) 
        }
    # the`try / except` behaves like second defence if the above logic didn't catch all the messy data
    except (TypeError, ValueError):
        return None   # skip rows with missing/bad data if they are not caught in the above
````

## 4. Create clean dataset

````python
# use list comprehension to clean each row of record
clean = [parse_record(r) for r in raw_data]
````

## 5. Further analysis

````python
# --- Step 2: Filter ---
# list comprehension with condition to filter
high_margin = [r for r in clean if r["margin"] >= 0.30]
# \n insert a blank row after output
print(f"High-margin rows: {len(high_margin)}\n")

# --- Step 3: Aggregate by region ---
# set comprehension for unique record (aka distinct category)
regions = {r["region"] for r in clean}

# sorted(set) to order regions alphebetically
for region in sorted(regions):
    # list comprehension with condition to filter all rows per region
    rows = [r for r in clean if r["region"] == region]
    # generator expression to sum revenue & profit for the region's rows
    # A generator expression is a shorter and more compact way to create a generator without using a function or yield. It looks similar to a list comprehension, but uses parentheses () instead of brackets [].
    total_rev = sum(r["revenue"] for r in rows)
    total_profit = sum(r["profit"] for r in rows)
    # print formatted results
    print(f"{region:8s}  Revenue: {total_rev:>10,.0f}  Profit: {total_profit:>10,.0f}")

# --- Step 4: Top product by revenue ---
# use lambda to sort list by "revenue" in decending order
# slice index 0, which is the highest "revenue"
top = sorted(clean, key=lambda r: r["revenue"], reverse=True)[0]
# \n insert a blank row before output
print(f"\nTop product: {top['product']} in {top['region']} — {top['revenue']:,.0f}")
````
