```python
import pandas as pd
```

```python
#Load CSV files
df_c = pd.read_csv("customers.csv")
df_da = pd.read_csv("delivery_addresses.csv")
df_ol = pd.read_csv("order_line.csv")
df_il = pd.read_csv("invoice_line.csv")
df_d = pd.read_csv("deliveries.csv")
```

```python
df_c.head()
```

```python
df = df_ol.merge(
    df_il,
    on=["order_id", "product_id"],
    how="outer",
    suffixes=("_order", "_invoice")
)
```

```python
df["quantity_diff"] = df["quantity_invoice"] - df["quantity_order"]
```

```python
df["quantity_issue"] = df["quantity_diff"].apply(
    lambda x: "OK" if x == 0 else ("Over-invoiced" if x > 0 else "Under-invoiced")
)
```
